## Vector operations

```python
import numpy as np

u=np.array([1,2,3,4])
v=np.array([3,4,5,6])
```

```python
print(u+v)
print(u*v)
```

```python
v.shape[0]
```

```python
##vector multiplication

def vector_multiplication (u,v):
    assert u.shape[0]==v.shape[0]
    n=u.shape[0]##it could nbe v as well both have same shape
    result=0.0
    for i in range(n):
        result=result+u[i]*v[i]
    return result
```

```python
vector_multiplication(u,v)
```

```python
## matrix vector multiplication
a=np.array([[2,4,5,6],[1,2,1,2],[3,1,2,1]])
b=np.array([1,0.5,2,1])

print(a.shape)
print(b.shape)
```
