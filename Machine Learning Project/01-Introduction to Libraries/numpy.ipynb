## Random Arrays


```python
np.random.seed(2)## setting the seed to reproduce same numbers
np.random.rand(5,2)
```

```python
np.random.seed(2)
np.random.randn(5,2)
```

```python
##random integers
np.random.seed(2)
np.random.randint(low=0,high=20,size=(5,2))
```
