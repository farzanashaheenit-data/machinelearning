### Pandas Intro and Functions

```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/datasets/master/laptops.csv")

df.head(10)
```

```python
pd.__version__
```

```python
#ROWS & COLUMNS
df.shape
```

```python
##index

df.index

##accesing elements in df use loc funtion 
df.loc[[1,2,8]] ### loc is label based incase index has labels else it would work in our case we have numbers
df.loc[1:3]

df.iloc[1:2] ## iloc is used when we are accessing elements if index is integer
```

#### string operations

```python
df['Laptop'].str.lower()
df['Laptop'].str.upper()
df['Laptop'].str.capitalize()
```

##### sumarizing operators

```python
print(df['Final Price'].max())
print(df['Screen'].median())
print(round(df['Final Price'].sum(),2))
```
